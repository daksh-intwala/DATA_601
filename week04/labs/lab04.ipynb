### 4. Which plane (tailnum) has the worst on-time record?


```python
flights['max_late']= flights[['dep_delay','arr_delay']].max(axis=1)
```

```python
flights.head()
```

#### Taking average and grouping on-time of each tailnum plane from max_value = avg (time - scheduled time)
     Each Plane has several flights so this code block averages out 'avg delay time' from all flights of single plane.

```python
avg_ontime_per_flight=[]
time = []
for key, tail in flights.groupby(by='tailnum'):
    avg_ontime_per_flight.append((key,tail['max_late'].mean()))
    time.append(tail['max_late'].mean())
```

```python
avg_ontime_per_flight
```

```python
for i in range(len(avg_ontime_per_flight)):
    if avg_ontime_per_flight[i][1]==max(time):
        print(avg_ontime_per_flight[i][0],avg_ontime_per_flight[i][1])
```

### 5. For each plane, count the number of flights before the first delay of greater than 1 hour.

```python
flights.head()
```

```python
planes = flights.groupby(by='tailnum')
```

```python
ls=[]
for key,values in planes:
    ls.append((key, values[(values.dep_delay < 60) & (values.arr_delay < 60)].count()[0]))
```

```python
ls
```

This lsit 'ls' displays the count of flights for each plane with less than 1 hour delay.
