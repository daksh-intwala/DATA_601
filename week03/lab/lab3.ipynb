```python
flights.head(5)
```

## Question 1 : Had an arrival delay of two or more hours.

```python
flights[flights['arr_delay'] >=2]
```

## Question 2 : Flew to Houston (IAH or HOU)

```python
flights[(flights['dest'] == 'IAH' ) | (flights['dest'] == 'HOU' )]
```

## Question 3 : Were operated by American, Delta

```python
flights[(flights['carrier'] == 'AA' ) | (flights['carrier'] == 'DL' )]
```

## Quetion 4 : How many values are missing in `dep_time`?

```python
flights['dep_time'].isnull().sum()
```

## Question 5 : Sort `flight` to find fastest flight.

```python
flights['flight_time1']= (flights['hour']) + (flights['minute']/60)
```

```python
flights['flight_time1'].argmin()
```

```python
flights.iloc[275945]
```

### Flight 1632 is the Fastest Flight in terms of *'travel time'* with total flight time 1hr and 6min

```python

#By Sorting
flights.sort_values(by=('flight_time1'))
```

#### BY Velocity

```python
#***EXTRA***
#In terms of velocity

flights['velocity']= (flights['distance'])/((flights['flight_time1']))
flights[['flight','velocity']].loc[0]
                    
```

## Question 6 : Which flights travelled the shortest?

```python
flights.sort_values(by='distance', ascending=True).iloc[0]
```
