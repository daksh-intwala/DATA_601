## Lab Part1

Please choose appropriate graph and visualize the followings with correct/appropriate labels

* Height only
* Points only
* Height vs Points

```python
import pandas as pd

nba_players = pd.read_csv('https://raw.githubusercontent.com/msaricaumbc/DS_data/master/all_seasons.csv', index_col = 0)

height = nba_players['player_height']
points = nba_players['pts']
```

```python
import matplotlib.pyplot as plt
```

### Visualizing Height
#### Histogram Plot

```python
plt.hist(height, alpha=0.6, color='orange')

plt.xlabel("Heights")
plt.ylabel("Number of Players")

font={'size': 14, 'family':'serif'}

plt.title("Histagram Plot for NBA player's height", pad=15, fontdict=font)
plt.grid()
plt.show()
```

#### Box Plot

```python
plt.boxplot(height)

plt.ylabel("Height")

plt.title("Box Plot for NBA player's height", pad=15, fontdict=font)
plt.show()
```

### Visualizing Points
#### Histogram Plot

```python
plt.hist(points, alpha=0.6, color='green')

plt.xlabel("Points")
plt.ylabel("Number of Players")

plt.title("Histagram Plot for NBA player's points", pad=15, fontdict=font)
plt.grid()
plt.show()
```

#### Box Plot

```python
plt.boxplot(points)

plt.ylabel("Points")

plt.title("Box Plot for NBA player's points", pad=15, fontdict=font)
plt.show()
```

### Visualizing Height Vs. Points
#### Scatter Plot

```python
plt.scatter(height, points)

plt.xlabel('Height of Players')
plt.ylabel('Points')

plt.title('Height of NBA Player vs Points',pad=15, fontdict=font)

plt.draw()
```

## Lab Part2
Please visualize year vs score with correct/appropriate labels

```python
bowling_data = pd.read_csv("https://raw.githubusercontent.com/msaricaumbc/DS_data/master/bowling_stats.csv",
                            header=None,
                            names=['year','city','state','count1','count2'])
def merge_columns(row):
    if pd.isna(row['count2']):
        return row['count1']
    else:
        return row['count1']*1000+row['count2']
    
bowling_data['total']=bowling_data.apply(merge_columns,axis=1)

year = bowling_data['year']
score = bowling_data['total']
```

```python
plot_1= plt.figure(figsize=(35,10))

plt.scatter(year, score, alpha=1, color='red')

font2= {'size':35, 'family':'serif'}
font3={'size':25}
plt.xlabel('Year', fontdict=font3)
plt.xticks(rotation=45)
plt.ylabel('Score', fontdict=font3)
plt.title('Year vs. Score',pad=15, fontdict=font2)
plt.grid()
plt.draw()
```
