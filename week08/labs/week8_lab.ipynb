__Your Task__

- In  `Q12_Part_4_TEXT` they ask the tools the participants are using day to day. However as you might observe the answers are not standardized. 

- Find all the answers with 'jupyter' mentioned in it.

- Find all the participant who are Python users.

- Find all R users. 

```python
df
```

### Find all the answers with 'jupyter' mentioned in it.

```python
df.fillna('', inplace=True)
```

```python
jp=df[df['Q12_Part_4_TEXT'].str.lower().str.contains('jupyter|jup|puter|piter|notebook|pyter')]['Q12_Part_4_TEXT']
```

```python
for i in jp:
    print(i)
```

```python
len(jp)
```

### Find all the participant who are Python users.

```python
python = df[df['Q12_Part_4_TEXT'].str.lower().str.contains('pyth|python|pyton|charm|colab|pandas|sckit|scikit')]['Q12_Part_4_TEXT']
```

```python
for i in python:
    print(i)

```

```python
len(python)
```

### Find all R users.

```python
r = df[df['Q12_Part_4_TEXT'].str.lower().str.contains('[ ]r[ ,]|^r$|r[ -][s]|r-studio|rs|^r[, ]|[ ]r$')]['Q12_Part_4_TEXT']
```

```python
for i in r:
    print(i)

```

```python
len(r)
```
