## Loading dataset from week's lab

```python
import pandas as pd

countries = pd.read_csv('/Users/dakshintwala/github_temp/DATA_601/Week11/labs/countries.csv')
```

## Scraping Data

```python
tables = pd.read_html('https://en.wikipedia.org/wiki/List_of_countries_by_past_and_projected_GDP_(nominal)')
```

```python
tables[5].head(5)
```

```python
countries.head(5)
```

```python
gdp_data = tables[5]
```

```python
gdp_data = gdp_data.rename(columns = {'Country (or dependent territory)':'country'})
```

```python
gdp_data.shape
```

```python
countries.shape
```

```python
merged_df = pd.merge(countries, gdp_data[['country','2022','2023','2024','2025','2026','2027']], on='country', how = 'left')
```

```python
merged_df.head(5)
```

```python
merged_df['avg_increase(%)'] = (merged_df['2027'] - merged_df['2022'])*100 / merged_df['2022']
```

```python
final_output = merged_df.sort_values(by='avg_increase(%)', ascending = False)
```
