```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

diamonds = sns.load_dataset("diamonds")

diamonds.head()
```

#### 1.

- Plot the scatter plot of price and carat. Add clarity into the plot as hue. 

- Do you see any pattern? Is it parallel to your expectation?

```python
_=sns.scatterplot(x=diamonds.carat, 
                  y=diamonds.price, 
                  hue=diamonds.clarity, 
                  alpha=1)
sns.set(rc = {'figure.figsize':(25,15)})
plt.title( "Carat Vs. Price - Scatter Plot", fontdict={"fontsize":30, "style":'oblique'})
plt.xlabel("Carat",fontdict={"fontsize":25})
plt.ylabel("Price",fontdict={"fontsize":25})
# plt.figure(figsize=(50,30))
plt.show()
```

Patterns Recognized:
1. Violet and pink are seen maximum which represents VS2 and SI2 calrity diamonds.
2. As we mover towards higher carat diamonds, average diamond price
   increases above 3 carat, having only high prices and all possessing I1 clarity.
3. Diamonds with Carats within range of 1 to 2/2.5 has almost all price ranges availble from highest        price to medium price and all clarity levels seem to be present.
4. Diamonds with less than 1 carat has lowest avg price. 
5. IF clarity diamonds seem to be rarely present in diamonds with more than 1.5 carat.
6. Some carat such as 0.5,1,1.5,2,etc seem to have maximum datapoints indicating that there might be an      accepted standard of carat in diamonds which is preferable/desired.
7. It is according to my expectation to have few diamonds of higher prices and most diamonds of mid-range price in the market.


#### 2.

- Consider `clarity` and `color` is there any pattern in their heatmap?

```python
plot = sns.displot(diamonds, 
                   x="clarity", 
                   y="color", 
                   cbar = True, 
                   color='red')
plt.title( "Clarity Vs. Color - Heatmap", fontdict={"fontsize":16, "style":'oblique'})
plt.xlabel("Clarity",fontdict={"fontsize":14})
plt.ylabel("Color",fontdict={"fontsize":14})
plt.show()
```

Pattern Recognized:
1. The maximum records are with these specifications: Clarity=['VS1','VS2','SI1'] & Color =            ['G','E','H','F'] mainly above 2000 diamond counts and some between 1500-2000.
2. Diamonds with I1 clarity are very few in count (741) in comparison to total. They may not exist.
3. IF clarity diamonds are too very less, they may not be so popular(need to verify with an expert).
4. WS1, WS2, VS1, VS2, SI1, SI2 seem to be popular clarity in diamonds with main density of heatmap    focused on their part.

#### 3.

- Use `carat` variable as as the size of the points in scatter plot in which x-axis is the `depth` and `price` on the y-axis.

```python
__=sns.scatterplot(data=diamonds, 
                   x="depth", 
                   y="price", 
                   size="carat",
                   alpha=0.5,
                  )
sns.color_palette()
sns.set(rc = {'figure.figsize':(30,15)})
plt.title("Depth Vs. Price - Scatter Plot categorized by carat", fontdict={"fontsize":25, "style":'oblique'})
plt.xlabel("Depth",fontdict={"fontsize":15})
plt.ylabel("Price",fontdict={"fontsize":15})
plt.show()
```

1. Most of the datapoints lie in between depth of 55-70 which shows that this range of depth might be            desirable in the market. 

2. The outliers with depth less than 55 and depth more than 70 seem to be of smaller carat 1,2 according to      the plot.
